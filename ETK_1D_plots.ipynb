```python
import matplotlib.pyplot as plt
import numpy as np
from kuibit.simdir import SimDir
from kuibit.grid_data import UniformGrid
```

## **Sod Shock Tube Problem**

```python
#initial data from the Exact Riemann solver of https://www.brunogiacomazzo.org/?page_id=395
exact = np.loadtxt("Sod_exact.dat")
```

```python
x_exact=exact[:,0]
rho_exact=exact[:,1]
press_exact=exact[:,2] #note: this is pgas+pmag
vx_exact=exact[:,3]
vy_exact=exact[:,4]
vz_exact=exact[:,5]
By_exact=exact[:,6]
Bz_exact=exact[:,7]
```

```python
fig, axs = plt.subplots(1, 3, figsize=(15, 4))
fig.suptitle("Exact solution")

axs[0].plot(x_exact, rho_exact, label="rho (exact)", color='teal')
axs[0].set_xlim(-0.5, 0.5)
axs[0].set_title("Density")
axs[0].legend()
axs[0].grid(True)

axs[1].plot(x_exact, press_exact, label="press (exact)", color='teal')
axs[1].set_xlim(-0.5, 0.5)
axs[1].set_title("Pressure")
axs[1].legend()
axs[1].grid(True)

axs[2].plot(x_exact, vx_exact, label="vx (exact)", color='teal')
axs[2].set_xlim(-0.5, 0.5)
axs[2].set_title("vx")
axs[2].legend()
axs[2].grid(True)

plt.tight_layout()
plt.show()
```

```python
gf_100 = SimDir("./Sod_1d_100/").gf
print(gf_100)

gf_200= SimDir("./Sod_1d_200/").gf
print(gf_200)

gf_400= SimDir("./Sod_1d_400/").gf
print(gf_400)

gf_800= SimDir("./Sod_1d_800/").gf
print(gf_800)

gf_1000= SimDir("./Sod_1d_1000/").gf
print(gf_1000)


```

```python
vars1d_100=gf_100.x
print(vars1d_100)
rho_100=vars1d_100.fields.rho
press_100=vars1d_100.fields.press
vx_100=vars1d_100.fields.vel[0]


vars1d_200=gf_200.x
print(vars1d_200)
rho_200=vars1d_200.fields.rho
press_200=vars1d_200.fields.press
vx_200=vars1d_200.fields.vel[0]

vars1d_400=gf_400.x
print(vars1d_400)
rho_400=vars1d_400.fields.rho
press_400=vars1d_400.fields.press
vx_400=vars1d_400.fields.vel[0]

vars1d_800=gf_800.x
print(vars1d_800)
rho_800=vars1d_800.fields.rho
press_800=vars1d_800.fields.press
vx_800=vars1d_800.fields.vel[0]

vars1d_1000=gf_1000.x
print(vars1d_1000)
rho_1000=vars1d_1000.fields.rho
press_1000=vars1d_1000.fields.press
vx_1000=vars1d_1000.fields.vel[0]

```

```python
small_grid = UniformGrid([101], x0=[-0.45], x1=[0.45])
```

```python
import seaborn as sns
print(rho_200.iterations)

colors = sns.color_palette("Set2", 3)

rho_initial = rho_200.read_on_grid(0, small_grid)
press_initial = press_200.read_on_grid(0, small_grid)
vx_initial = vx_200.read_on_grid(0, small_grid)


# Initial data
fig, axs = plt.subplots(1, 3, figsize=(15, 4))
fig.suptitle("Initial conditions (t=0)") 
axs[0].plot(*rho_initial.coordinates_meshgrid(), rho_initial.data_xyz, label="rho (initial)", color=colors[0])
axs[0].set_xlim(-0.5, 0.5)
axs[0].set_title("Density")
axs[0].legend()
axs[0].grid(True)  
axs[1].plot(*press_initial.coordinates_meshgrid(), press_initial.data_xyz, label="press (initial)", color=colors[1])
axs[1].set_xlim(-0.5, 0.5)
axs[1].set_title("Pressure")
axs[1].legend()
axs[1].grid(True)
axs[2].plot(*press_initial.coordinates_meshgrid(), vx_initial.data_xyz, label="vx (initial)", color=colors[2])
axs[2].set_xlim(-0.5, 0.5)
axs[2].set_title("vx")
axs[2].legend()
axs[2].grid(True)   
```

```python
import seaborn as sns

resolutions = [100, 200, 400, 800, 1000]
datasets = {
    'rho': [rho_100, rho_200, rho_400, rho_800, rho_1000],
    'press': [press_100, press_200, press_400, press_800, press_1000],
    'vx': [vx_100, vx_200, vx_400, vx_800, vx_1000]
}

colors = sns.color_palette("Set2", 3)

exact_solutions = {
    'rho': {'x': x_exact, 'y': rho_exact},
    'press': {'x': x_exact, 'y': press_exact},
    'vx': {'x': x_exact, 'y': vx_exact}
}

final_data = {}
for var_name, var_datasets in datasets.items():
    final_data[var_name] = []
    for i, dataset in enumerate(var_datasets):
        final_val = dataset.read_on_grid(dataset.iterations[-1], small_grid)
        final_data[var_name].append(final_val)

fig, axs = plt.subplots(3, 5, figsize=(25, 12))
fig.suptitle("Variables at final t = 0.4", fontsize=16)

var_info = {
    'rho': {'name': 'Density', 'color': colors[0]},
    'press': {'name': 'Pressure', 'color': colors[1]},
    'vx': {'name': 'Velocity X', 'color': colors[2]}
}

for row, (var_name, info) in enumerate(var_info.items()):
    for col, resolution in enumerate(resolutions):
        ax = axs[row, col]

        # exact solution in grey
        exact = exact_solutions[var_name]
        ax.plot(exact['x'], exact['y'], 
                label=f"{var_name} (exact)", 
                color='grey', linewidth=2, alpha=0.7)
        
        final_grid_data = final_data[var_name][col]
        ax.plot(*final_grid_data.coordinates_meshgrid(), final_grid_data.data_xyz, 
                label=f"{var_name} ({1/resolution})", 
                color=info['color'], linewidth=1.5)
        
        ax.set_xlim(-0.5, 0.5)
        ax.set_title(f"{info['name']} at final time ({resolution})")
        ax.legend()
        ax.grid(True)

#plt.tight_layout()
plt.show()
```

```python
fig, axs = plt.subplots(5, 3, figsize=(18, 20))
fig.suptitle("Variables at final t = 0.4", fontsize=16)

var_names = ['rho', 'press', 'vx']
var_titles = ['Density', 'Pressure', 'Velocity X']

for row, resolution in enumerate(resolutions):
    for col, var_name in enumerate(var_names):
        ax = axs[row, col]
        exact = exact_solutions[var_name]
        ax.plot(exact['x'], exact['y'], label=f"{var_name} (exact)", color='grey', linewidth=2, alpha=0.7)
        final_grid_data = final_data[var_name][row]
        ax.plot(*final_grid_data.coordinates_meshgrid(), final_grid_data.data_xyz, label=f"{var_name} (dx = {1/resolution})", color=colors[col], linewidth=1.5)
        ax.set_xlim(-0.5, 0.5)
        ax.set_title(f"{var_titles[col]} (dx = {1/ resolution})")
        ax.legend()
        ax.grid(True)

plt.tight_layout()
plt.show()
```
